```python
import pyxdf
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(precision=7,formatter={'float': '{: 0.5f}'.format})
```

```python
#get combined data
file_folder=r'D:\projects\testing_the_eeg_room_ido_and_noor\sub-P001\ses-S001\eeg'
file_loc=r'\sub-P001_ses-S001_task-Default_run-002_eeg.xdf'
data_combined, header = pyxdf.load_xdf(file_folder+file_loc)
```

```python
#check data combined to see which stream appears where
tobii=data_combined[0]
lsl_trigger=data_combined[1]
eeg_trigger=data_combined[2]
eeg_signal=data_combined[3]
et_trigger=data_combined[4]
```

```python
#plot eeg latency compared to lsl
eeg_lsl=eeg_trigger['time_stamps']-lsl_trigger['time_stamps'][:-1] #remove last data point from lsl trigger
plt.plot(eeg_lsl)
plt.title("EEG-LSL triggers latency")
plt.xlabel("Trial")
plt.ylabel("Time")
plt.show()
```

```python
#plot et latency compared to lsl
et_187=et_trigger['time_stamps'][np.where(et_trigger['time_series'][:,1]==187)] #take only 187 triggers from et
et_lsl=et_187[2:]-lsl_trigger['time_stamps'][:-1][2:]
plt.plot(et_lsl)
plt.title("ET-LSL triggers latency")
plt.xlabel("Trial")
plt.ylabel("Time")
plt.show()
```
