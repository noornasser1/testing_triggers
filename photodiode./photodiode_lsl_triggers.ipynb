```python
import pyxdf
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(precision=7,formatter={'float': '{: 0.5f}'.format})
```

```python
#get combined data
file_folder=r'D:\projects\testing_the_eeg_room_ido_and_noor\sub-P001\ses-S001\eeg'
file_loc=r'\sub-P001_ses-S001_task-Default_run-006_eeg.xdf'
data_combined, header = pyxdf.load_xdf(file_folder+file_loc)
```

```python
tobii=data_combined[0]
eeg_trigger=data_combined[1]
eeg_signal=data_combined[2]
lsl_trigger=data_combined[3]
```

```python
plt.plot(eeg_signal['time_stamps'][:18000]-eeg_signal['time_stamps'][0],eeg_signal['time_series'][:18000,0])
plt.plot(lsl_trigger['time_stamps'][:12]-eeg_signal['time_stamps'][0], np.zeros(12)+0.05, 'x')
plt.title("Photodiode signal and lsl triggers")
plt.xlabel("Time")
plt.ylabel("Signal")
plt.show()
```
